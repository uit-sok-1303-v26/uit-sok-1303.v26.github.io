### 1. Markedsetterspørsel

En konsument disponerer inntekt, og velger mellom mange goder og tjenester. Hvor mye av hvert gode som etterpørres avhenger av
- konsumentens preferanser
- prisen på produktet
- konsumentens inntekt
- prisen på andre produkter/tjenester
- offentlig regulering.

Konsumentens etterspørsel etter et produkt viser hva hen er villig til å betale for en ekstra enhet, dvs *marginal betalingsvillighet*. Den individuelle etterspørselen kan uttrykkes som $$ Q_{i}^E=a_{i}-b_{i}P.$$

![Bilde](Picture10.png)

Ved å summere konsumentenes etterspørselkurve får vi markedets etterspørselskurve $$ Q^E=a-bP.$$

```python
# import pakkene som vi trenger for å tegne en figur
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Definer Lineær etterspørsel til konsument 1 og 2
# cf fig 3.2 i boka
# x=a - b*p
#Vi setter a_1 = a_2=100, b_1=1, b_2=2



def demand_1(p):
    return 100 - p

def demand_2(p):
    return 100 - 2*p

def m_demand(p):
    w=demand_1(p)
    z=demand_2(p)
    sum = (z<0)*w + (z>=0)*(z+w)

        
    return sum
```

```python
# Tegne individuell etterspørsel og den horisontale summen

p_num = np.linspace(0,110,100)

fig, axs = plt.subplots(1,3, sharey=True, figsize=(20,5))

# plott av linjene

axs[0].plot(demand_1(p_num),p_num,label='Etterspørsel kons 1')
axs[1].plot(demand_2(p_num),p_num,label='Etterspørsel kons 2')
axs[2].plot(m_demand(p_num),p_num,label='Markedsetterspørsel')

# bruk løkke for å tegne aksene

for i in range(3):

    axs[i].set_ylabel('Pris (P)', loc ='top', rotation = 0)
    axs[i].set_xlabel('Kvantum (Q)', loc = 'right')
    axs[i].set(ylim=(0,110),xlim=(0,210))
    axs[i].tick_params(labelleft=True)
    axs[i].spines['top'].set_color('none')
    axs[i].spines['right'].set_color('none')
    axs[i].legend(loc='best', frameon=False)

# gi en passende tittel

fig.suptitle('Utledning av etterspørsel i markedet')

plt.show()
```

#### 1.1 Elastisitetsbegrepet

Elastisitet gir oss et mål på hvor følsomt tilbud og etterspørsel er for prisendringer. Vi har tidligere tegnet disse funksjonene som en rett linje, og helningen til linjen forteller oss noe om hvordan mengde tilbudt/etterspurt reagerer på prisendringer. Tenk at vi har invers etterspørsel etter laks i Norge gitt ved $P = 300 - 0.25Q$. Her er prisen målt i NOK og mengde i kilogram. Vi ser at for å selge en ekstra kilo med laks i Norge må prisen reduseres med 25 øre. I Spania er etterspørsel etter laks gitt ved $q = 500 - 0.3Q$, og q er nå prisen i EUR. For å selge en ekstra kilo laks til det spanske markedet må prisen reduseres med 30 cent. For å sammenlikne mellom Norge og Spania må vi ha informasjon om valutakursen mellom NOK og EUR. Dette eksempelet illustrerer at det å bruke helningen til etterspørsel/tilbud for å måle prisfølsomhet er avhengig av måleenheten for verdi. Sammenlikningen vil også påvirkes av hvordan vil måler kvantum (kilo, ton osv).     

Elastisitet er målt i prosent og er da uavhengig av måleenheten for verdi og mengde. Etterspørselselastisiteten måler hvor mange prosent etterspørselen endres når prisen øker med 1%. Tenk at vi har etterspørsel $Q^E(P)$. En liten endring i etterspørsel kan vi kalle $dQ^E$, slik at en prosentvis endring i etterspørsel er
$$prosentvis\; endring \;i \; etterspørsel = \frac{dQ^E}{Q^E}\tag{1}.$$

La $dP$ vise en liten endring i pris, og vi kan skrive den prosentvise endringen i pris som 
$$prosentvis\; endring \;i \; pris = \frac{dP}{P}\tag{2}.$$

Etterspørselens priselastisitet er gitt ved $\epsilon^E$ som er den prosentvise endringen i kvantum etterspurt delt på den prosentvise endringen i pris.
$$\epsilon^E = \frac{prosentvis\; endring \;i \; etterspørsel}{prosentvis\; endring \;i \; pris}.$$
Tenk på et talleksempel før vi utleder dette formelt. Vi undersøker etterspørsel i et marked og har funnet at etterspørsel har nylig falt med 6%; i samme periode har prisen økt med 3%. Deler vi den ene på den andre så finner vi at -6%/3% = -2% (minus ettersom etterspørselen har falt). Tolkningen nå er når prisen øker med 1% så faller etterspørsel med 2%. Det er nettopp dette som etterspørselens priselastisitet måler!

Ved å bruke likning (1) og (2) kan vi skrive
$$\epsilon^E = \frac{\frac{dQ^E}{Q^E}}{\frac{dP}{P}}.$$
Her deles en brøk på en annen brøk som vi kan forenkle som følger
$$\epsilon^E = \frac{\frac{dQ^E}{Q^E}}{\frac{dP}{P}}=\frac{dQ^E}{Q^E}\frac{P}{dP}.$$

Som vi kan skrive om som:
$$\epsilon^E = \frac{\frac{dQ^E}{Q^E}}{\frac{dP}{P}}=\frac{dQ^E}{Q^E}\frac{P}{dP}=\frac{dQ^E}{dP}\frac{P}{Q^E}.$$

Legg merke til at $\frac{dQ^E}{dP}$ er helningen til den direkte etterspørselsfunksjonen $Q^E$. Ettersom etterspørselen heller nedover er $\frac{dQ^E}{dP} < 0$ og dermed er $\epsilon^E < 0.$

#### Eksempel med lineær etterspørsel

La oss tenke oss følgende lineære etterspørselsfunksjon:    

$Q^E = 500- 5P$ i punkt $P=50, Q^E=250.$

La oss tegne dette først.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def demand(P):
    return 500-5*P

p_num=np.linspace(0.1,99.9,100)

#lag en akse

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris (P)', loc='top', rotation = 0)
                ax.set_xlabel('Kvantum (Q)', loc='right')
                ax.set(xlim=(0,400))
                ax.set(ylim=(0,150))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonen

ax.plot(demand(p_num),p_num, label='Etterspørsel')


# tittel
ax.set_title('Lineær etterspørsel')

#vis navnene:
ax.legend(loc='best',frameon=False);
```

Vi kan regne ut et analytisk uttrykk for etterspørselens priselastisitet.

```python
import sympy as sp

Q, P = sp.symbols('Q, P', real=True, positive=True)

def Q_d(P):
    return 500-5*P

el_d = sp.diff(Q_d(P), P)*P/Q


el_d
```

Deretter kan vi sette inn tall for pris og kvantum.

```python
el_d.subs([(Q,250),(P,50)])
```

Da er $\epsilon^E=-1$ i punkt P=50, Q= 250. Målt fra dette punktet på etterspørselskurven vil en 1% økning i pris føre til en 1% reduksjon i kvantum etterspurt.

```python
ax.plot ([0, 250], [50, 50], linestyle='dashed', color='black')
ax.plot ([250, 250], [0, 50], linestyle='dashed', color='black')
ax.annotate("ettersp.el=-1", xy=(250, 50), xytext=(275, 80), arrowprops={"arrowstyle":"->", "color":"gray"})
fig
```

### 2. Markedstilbudet

Produsenter kombinerer innsatsfaktorer for å produsere varer eller tjenester. Hvor mye de må ta for å dekke den neste enheten som produseres vil avhenge av
- produksjonskostnader
- teknologien
- offentlige reguleringer.

På samme vis som ovenfor kan vi summere opp tilbudet til hver produsent for å finne markedstilbudet $$ Q^T=-y+dP.$$

Et lineært eksempel er tegnet nedenfor.

```python
def tilbud(p):
    return -20 + 1.5*p

#lag en akse som vi skal bruke flere ganger

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris (P)', loc='top')
                ax.set_xlabel('Kvantum (Q)', loc='right')
                ax.set(xlim=(0,110))
                ax.set(ylim=(0,110))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

# plott funksjonen

ax.plot(tilbud(p_num),p_num, label='tilbud')


# tittel
ax.set_title('Markedets tilbud')

#vis navnene:
ax.legend(loc='best',frameon=False);
```

Tilbudet viser hva produsentene må ha for å dekke kostnaden av den siste enheten produsert, dvs produsentens grensekostnad.

#### 2.1 Eksempel på tilbusdelastitstet


Vi har en lineær tilbudsfunksjonen $X^S(p)$. Tilbudselastisiteten viser hvor mange prosent tilbudet øker når prisen går opp med 1%, og skrives som
$$\epsilon^T=\frac{dQ^T}{dp}\frac{P}{Q^T}.$$
Ettersom tilbudet heller oppover har vi $\frac{dQ^T}{dP}>0$ og dermed $\epsilon^T>0.$     
Ta et eksempel med $Q^T=-20+1.5P$. Vi kan regne ut at elastisiteten til tibudet er 
$$\epsilon^T=1.5 \frac{P}{Q^T}=1.5 \frac{P}{-20+1.5P}.$$

```python
def tilbud(p):
    return -20 + 1.5*p

def elast_t(p):
    return 1.5*p/(-20+1.5*p)

p_ny=np.linspace(14,100,100)

# # lag en akse med 2 paneler
def create_ax2():
                fig, axs = plt.subplots(1,2, sharey=True, figsize=(15,5))
                for i in range(2):
                    axs[i].set_ylabel('Pris (p)', loc='top', rotation = 0)
                    axs[i].tick_params(labelleft=True)
                    axs[i].spines['top'].set_color('none')
                    axs[i].spines['right'].set_color('none')
                return fig, axs

fig, axs = create_ax2()


axs[0].set(xlim=(0,200))
axs[1].set(ylim=(0,100))
axs[1].set(xlim=(0,3))
axs[0].set_xlabel('Kvantum (x)', loc='right')
axs[1].set_xlabel('Tilbudselastisitet', loc='right')

# plott funksjonenen

axs[0].plot(tilbud(p_ny),p_ny, label= 'Tilbud')
axs[1].plot(elast_t(p_ny),p_ny, label='Tilbudselastisitet')


# vis forskellige verdier i første panel


axs[0].annotate(r'$\epsilon^S faller$', xy=(130, 90), xytext=(55, 45), fontsize=15, arrowprops={"arrowstyle":"->", "color":"gray"})


#vis navnene:
axs[0].legend(loc='best',frameon=False)
axs[1].legend(loc='best',frameon=False)


# tittel 
axs[0].set_title('Tilbud')
axs[1].set_title('Tilbudselastisitet')
fig.suptitle('Lineært tilbud');
```

### 3. Markedslikevekt
Markedlikevekten finner vi ved sette tilbud = etterspørsel:
$$ Q^E=Q^T.$$  $$ a-bP=-y+dP.$$

```python
import sympy as sp

P, Q,y,d,a,b = sp.symbols('P, Q,y,d,a,b', real=True, positive=True)


eq_T = sp.Eq(-y+d*P, Q)
eq_E = sp.Eq(a-b*P, Q)

sol_dict = sp.solve ((eq_T, eq_E), (P,Q))

# løsningen er et oppslag (dictionary)
# vi får hver løsning ved å bruke sol_dict[P] og sol_dict[Q]
sol_dict
```

### Eksempel

Ved å sette sammen tilbud og etterspørsel får vi en beskrivelse av begge sidene av markedet. Her tegnes et marked med følgende tilbud og etterspørsel:
$$ Q^T =-20 + 1.5P \;(tilbud)$$
$$ Q^E = 100 - 1.25P \;(etterspørsel).$$

```python
def etters(P):
    return 100 - 1.25*P

fig, ax = create_ax()

ax.plot(tilbud(p_num), p_num ,label='tilbud')
ax.plot(etters(p_num), p_num ,label='etterspørsel')

ax.plot ([0,tilbud(60)], [60,60], linestyle='dashed')
ax.text (30, 70, "Tilbudsoverskudd")

ax.plot ([0,etters(30)], [30,30], linestyle='dashed')
ax.text (25, 20, "Etterspørselsoverskudd")


ax.set_title('Markedsmekanismen')

ax.legend(loc='best', frameon=False);
```

Figuren viser at når prisen er "for høy" vil produsenter tilby mer enn kjøpere etterspør; er prisen "for lav" vil det være udekket etterspørsel. Det vil alltid være den korte siden av markedet som bestemmer hvor mye som blir omsatt ettersom det er frivillig å kjøpe/selge. I tilfeller med tilbudsoverskudd er omsatt mengde satt av etterspørsel. For å selge mer kan produsenter sette ned prisen litt (og fortsatt få dekket grensekostnaden). Med etterspørselsoverskudd er det tilbudet som begrenser omsetning i markedet. For å være villig til å produsere mer må selgere få bedre betalt, og det er konsumenter som er villige til å betale en høyere pris. Prismekanismen gjør at vi beveger oss mot punktet hvor tilbud og etterspørsel krysser. I dette punktet er markedet dekket - alle som vil kjøpe til rådende pris får kjøpe, og alle som er villig til å selge til denne prisen får solgt. Her er markedets likevekt.

Vi kan løse for denne likevekten. Vi leter etter ett punkt som er både på tilbuds- og etterspørselsfunksjonen, dvs vi må løse likningssettet for P og Q. 

```python
import sympy as sp

P, Q = sp.symbols('P, Q', real=True, positive=True)


eq_s = sp.Eq(-20+1.5*P, Q)
eq_d = sp.Eq(100-1.25*P, Q)

sol_dict = sp.solve ((eq_s, eq_d), (P,Q))

# løsningen er et oppslag (dictionary)
# vi får hver løsning ved å bruke sol_dict[P] og sol_dict[Q]
sol_dict
```

```python
fig, ax = create_ax()


ax.plot(tilbud(p_num), p_num ,label='tilbud')
ax.plot(etters(p_num), p_num ,label='etterspørsel')

# sett inn linjer for likevekten

ax.plot ([0,sol_dict[Q]], [sol_dict[P], sol_dict[P]], linestyle='dashed', color='black')
ax.plot ([sol_dict[Q], sol_dict[Q]], [0, sol_dict[P]], linestyle='dashed', color='black')


ax.set_title('Markedslikevekt')

ax.legend(loc='best', frameon=False);
```

### 4. Skift i tilbud og etterspørsel

Dersom prisen på varen/tjenesten som vi ser på endres beveger vi oss **langs** etterspørsels/tilbudskurven. Dersom andre faktorer endrer seg får vi et **skift** i tilbud eller etterspørsel. Skifter tilbudet til høyre (venstre) vil produsenter tilby mer (mindre) til enhver pris. Likeså for etterspørsel: et skift til høyre (venstre) innebærer en økning (reduksjon) i etterspørsel.
Dersom bedrifter tar i bruk en kostnadsbesparende teknologi, eller prisen på en eller flere innsatsfaktorer faller får vi følgende effekt på markedslikevekten:

```python
def tilbud_ny(P):
    return -5+1.5*P

fig, ax = create_ax()


ax.plot(tilbud(p_num), p_num ,label='tilbud')
ax.plot(tilbud_ny(p_num), p_num ,label='nytt tilbud')
ax.plot(etters(p_num), p_num ,label='etterspørsel')

ax.set_title('Markedslikevekt med økt tilbud')

ax.legend(loc='best', frameon=False);
```

Vi ser at dette fører til et fall i likevektsprisen og en økning i omsetning.

Tenk nå at konsumenter får en økning i inntekt eller at prisen på et alternativt gode øker. Da er det nærliggende å tro at etterspørsel etter godet som vi ser på øker:

```python
def etters_ny(P):
    return 125-1.25*P

fig, ax = create_ax()

ax.plot(tilbud(p_num), p_num ,label='tilbud')
ax.plot(etters_ny(p_num), p_num ,label='ny etterspørsel')
ax.plot(etters(p_num), p_num ,label='etterspørsel')

ax.set_title('Markedslikevekt med økt etterspørsel')

ax.legend(loc='best', frameon=False);
```

Vi ser at omsetning øker, men at prisen også øker i ette tilfellet.

### 5. Invers tilbud og etterspørsel

Legg merke til hvordan vi har definert tilbud og etterspørsel: kvantum tilbudt/etterspurt som en funksjon av pris $Q^E(P), Q^T(P)$. Når vi tegner disse funkjsonene legg merke til at vi har pris på den vertikale og kvantum på den horisontale aksen, dvs vi har tegnet en funksjon $P(Q^E), \;P(Q^T)$. Disse kalles invers etterspørsel og invers tilbud.

Se igjen på tilbud og etterspørsel som vi brukte tidligere:
$$ Q^T =-20 + 1.5P \;(tilbud)$$
$$ Q^E = 100 - 1.25P \;(etterspørsel).$$

Vi kan løse hver likning for p for å få invers tilbud og etterspørsel:

```python
# vi kalte disse likningene eq_s og eq_d

inv_s = sp.solve (eq_s, P)
inv_s
```

```python
inv_d = sp.solve (eq_d, P)
inv_d
```

Da kan vi skrive invers tilbud og etterspørsel som:
$$ P = 13\frac{1}{3}+\frac{2}{3}Q^T \;(invers \;tilbud)$$
$$ P = 80 - 0.8Q^E \;(invers \;etterspørsel).$$

Her ser vi tydelig at tilbudet treffer p-aksen på $P = 13\frac{1}{3}$, og har konstant helning $\frac{dT}{dQ^T} = \frac{2}{3}$. Etterspørsel treffer p-aksen på $P = 80$, og har konstant helning $\frac{dP}{dQ^E} = -0.8$.

Vi kan selvsagt plotte invers etterspørsel, men må passe på å sette funksjonen som vi plotter på y-plassen i plot kommandoen:

```python
def inv_s(x):
    return 13.33+(2/3)*x

def inv_d(x):
    return 80 - 0.8*x

x_num = np.linspace(0,110,100)

fig, ax = create_ax()

# legg merke til at vi nå definerer verdien på x
# for så å finne p-verdien gjennom de inverse funksjonene

ax.plot(x_num, inv_s(x_num), label='tilbud')
ax.plot(x_num, inv_d(x_num), label='etterspørsel')


ax.set_title('Figur ved bruk av invers tilbud og etterspørsel')

ax.legend(loc='best', frameon=False);
```
