## Bedriftens innteker.

Når bedriften selger én enhet på produktmarkedet får den en fast pris på $P$. Med $Q$ solgte enheter får bedriften en inntekt på $ PQ.$

Det vil si at gjennomsnittsinntekt per enhet solgt er $\frac{PQ}{Q}=P$, og grenseinntekten er $\frac{dPQ}{dQ}=P$.   

For å produsere goder eller tjenester må bedriften ansette produksjonsfaktorer, noe som gir bedriften en kostnad. Kostnaden kan være fast, dvs uavhengig av hvor mange enheter som produseres: $FC=rK_0.$ Produksjonsavhengige kostnader er variable kostnader $VC$ som avhenger av mengden produsert. Totale kostnader blir da summen av disse: $TC=FC+V.$         
Dersom bedriften produserer én enhet ekstra ut fra dagens produksjonsnivå vil økningen i kostnaden kjennetegnes av grensekostnaden (MC):

$$
\frac{dC}{dQ}=\frac{dVC}{dQ}
$$

Beløpet som det koster bedriften per enhet produsert er den totale gjennomsnittskostnaden ($ATC$), og dette er summen av faste gjennomsnittskostnader ($AFC$) og variable gjennomsnittskostnader ($AVC$):


$$
ATC = AFC+AVC = \frac{FC}{Q} +\frac{AC}{Q}
$$




### Profittmaksimering
På kort sikt vil minst én produksjonsfaktor være fast, og dette vil utgjøre en del av bedriftens faste kostnader. Dersom bedriften produserer et produkt ved å kombinere kapital og arbeidskraft, og har en fast mengde med kapital på kort sikt ($K_0$) vil $rK_0$ være en del av bedriftens faste kostnad. Bedriften kan endre produksjon kun ved å endre mengde arbeidskraft som er brukt i produksjonsprosessen.        
Vi kan skrive fortjenesten som en funksjon av produsert mengde $F(Q)$:

$$F(Q)=PQ -(VC+FC).\tag{1}$$

Vi kan se at fortjeneste er positiv dersom $PQ>VC+FC$, dvs

$$F(Q)>0 \Rightarrow P>\frac{VC_f+FC}{Q}={ATC}.\tag{2}$$

Prisen (gjennomsnittsinntekten) må være større enn gjennomsnittskostnaden for positiv fortjeneste. I tillegg til dette er vi ute etter verdien på $Q$ som maksmimerer $F(Q)$. Betingelsen for maksinal profitt er

$$\frac{dF(Q)}{dQ}=P-\frac{dVC}{dQ}=0\Rightarrow P=MC. \tag{3}$$

Med ord: prisen (her grenseinntekten) må være lik grensekostnaden for et maksimum. Dersom $P>MC$ kan bedriften øke fortjeneste ved å selge en ekstra enhet, og produksjon kan ikke ha vært optimal. Dersom $P<MC$ har bedriften tapt penger på den siste enheten solgt, og kunne ha redusert produksjon for å øke fortjeneste. Da kan ikke denne produksjonen ha vært optimal. 


### Profittmaksimering og lønnsomhet

Vi har følgende produktfunksjon:


\begin{equation}
    Q = Q(K,L)=L^{0.5}K_0^{0.5}\tag{4}
\end{equation}

Fra forelesning Kapittel 6, matteboks 6.1 vet vi at grensekostnaden er gitt ved 

$$MC=\frac{dV}{dQ}=\frac{2wQ}{K_0}+z\tag{5}$$

Profittmaksimum der $P=MC$, som blir 

$$P=\frac{2wQ}{K_0}+z\tag{6}$$

```python
# Optimal profitt
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np


Q, K, K_0, L, P, w,r,z = sp.symbols('Q K K_0 L P w r z', real=True, positive=True)

def F(Q):
    return (P-(2*w*Q)/K_0)-z
F(Q)
```

```python
# Optimalt kvantum
Q_sol=sp.solve (F(Q),Q)[0]
Q_sol
```

## 7.2 Superprofitt i lakseoppdrett

Du er utnevnt som sekretær i et utvalg som skal se på beskatning i havbruket, og har fått i oppdrag av uvalgslederen å bruke mikroøkonomisk teori til å si noe om lønnsomheten i næringen. Anta at lakseproduksjonen er gitt ved produktfunksjonen $Q = K^{0.5}S^{0.5}$, der $K$ er naturkapital og $S$ er smolt (altså laksebabyer) som oppdretterne kjøper inn og videreforedler til ferdig laks.

Naturen er en gitt faktor, et konsesjonsområde, hvor både størrelsen på anlegget og antall i næringen er regulert av staten. Anta at størrelsen på hver konsesjon er gitt ved $K_0 = 1$. Tidligere ble konsesjonene delt ut gratis, mens produsentene i de siste årene har måttet betale for dem. Lakseprodusent $A$ var tidlig ute og fikk konsesjonen gratis, $r_A = 0$, mens $B$ kom til senere og måtte betale en pris $r_B = 1$. Prisen på smolt er lik for begge produsentene og gitt ved $s = 1$.

a) Regn ut kostnadene (grensekostnader, gjennomsnittlige variable kostnader, gjennomsnittlige totale kostnader) for produsent $A$ og $B$ og illustrer.

Starter med å skrive inn Cobb-Douglas-produktfunksjonen.

```python
S,K_0,w,r,Q,C,z, L,K = sp.symbols('S K_0 w r Q C z L K')

#k_0 = 1, w = 1, z = 0
K_0 = 1
w = 1
z = 0
# Definerer balansert cobb-douglas-funksjon
cobb_doug = sp.Eq(Q, K_0**0.5 * S**0.5)

cobb_doug
```

```python
# Løser for S, følger boken
S_losning = sp.solve(cobb_doug, S)[0]

S_losning
```

```python
# Fra matteboks 6.1 er kostnadene definert som
C = w*S_losning/K_0 + z*Q+ r*K_0

C
```

Grensekostnadene er lik for begge produsentene, og er gitt ved $MC = \frac{dC}{dQ}$.

```python
# Finner grensekostnaden ved å derivere C med hensyn på Q
MC = sp.diff(C, Q)

MC
```

Siden de er like blir da bare

$$
MC_A = MC_B = 2Q
$$


```python
# Definerer grensekostnaden som en funksjon for å plotte
def MC_plot(Q):
    return 2 * Q

# Fra matteboks 6.1, gjennomsnittlige variable kostnader
VC = w*Q/K_0 + z

# Fra matteboks 6.1, gjennomsnittlige faste kostnader
FC = r*K_0/Q


# Fra matteboks 6.1, gjennomsnittlige totale kostnader
ATC = VC + FC


# Definerer ATC som en funksjon for å plotte
def ATC_plot(Q,r):
    return Q + r/Q

# ATC_A, blir bare Q
ATC_plot(Q,0)
```

```python
# ATC_B, blir Q + 1/Q
ATC_B =ATC_plot(Q,1)
```

```python


# Deriverer for å finne minimum
ATC_B_derivert = sp.diff(ATC_B, Q)

ATC_B_derivert

# Løser for Q, henter ut løsning med positiv verdi
Q_losning = sp.solve(ATC_B_derivert, Q)[1]
Q_losning
```

```python
# Substituerer inn for K = 1 og Q = 1
ATC_B_losning = ATC_B.subs([(K,1), (Q,Q_losning)])
ATC_B_losning
```

Nå kan vi lage figuren

```python
# Figurkonfigurasjon
def create_ax():
    fig, ax = plt.subplots()
    ax.set_xlabel('Q', loc = "right")
    ax.set_ylabel('C', rotation=0, loc ='top')
    ax.set_ylim(0, 3.5)
    ax.set_xlim(0, 3.5)
    ax.set_xticks(np.arange(0, 4, 1))
    ax.set_yticks(np.arange(0, 4, 1))
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)
    
    return fig, ax
```

```python
Q_verdier = np.linspace(0.001, 3, 100)

fig, ax = create_ax()

# Tegner ATC og MC
ax.plot(Q_verdier, MC_plot(Q_verdier), label = "$MC_A$, $MC_B$", color = "cornflowerblue")
# r = 0 for ATC_A, r = 1 for ATC_B
ax.plot(Q_verdier, ATC_plot(Q_verdier,0), label = "$ATC_A$", color = "black")
ax.plot(Q_verdier, ATC_plot(Q_verdier,1), label = "$ATC_B$", color = "palevioletred")

# Tegner linjer for MC og ATC_B
ax.hlines(ATC_B_losning,0,Q_losning, linestyles = "--")
ax.vlines(Q_losning,0,ATC_B_losning, linestyles = "--")



ax.legend();
```

### Bedriftens tilpasning i faktormarkedet på kort sikt

La oss anta at bedriften produserer en vare av Arbeidskraft (L) og Kapital (K), og at kapitalmengden er fast på kort sikt $K=K_0$. Vi kan nå se på hva det er som bestemmer hvor mye arbeidskraft bedriften vil etterspørre for å maksimere fortjeneste på kort sikt.                  
         

1. Sett opp fortjenestesfunksjonen på kort sikt:     

$$
F(L) = P Q(L, K_0) - (rK_0 + w L)\tag{7}
$$

hvor $Q(L, K_0)$ er produktfunksjonen på kort sikt, $rK_0$ er faste kostnader og $wL$ er variable kostnader.         

2. Sett opp førsteordensbetingelsen:

$$
\frac{dF(Q)}{dL}=P\frac{Q(L, K_0)}{dN}-w=0\Rightarrow P\frac{Q(L, K_0)}{dN}=w \tag{8}
$$

Venstresiden er *verdien av grenseproduktet* som den siste enheten med arbeidskraft produserer, og høyresiden er *grensekostnaden* til den siste enheten med arbeidskraft.        

3. Løsningen til (8) er $L(K_0, w, P)$, dvs optimal mengde arbeidskraft etterspurt på kort sikt.  

####  Eksempel
En bedrift produserer en vare med produktfunksjon 

$$
Q = Q(K,L)=L^{0.5}K_0^{0.5}\tag{9}
$$

Videre setter vi $K_0=1, w=3, r=5$. 


```python
# Finn etterspørsel etter arbeid (L)

# Overskuddsmaksimering på kort sikt

import sympy as sp

K, L, P, w,r = sp.symbols('K L P w r', real=True, positive=True)

def prod(K,L):
    return L**(0.5)*K**(0.5)
def profitt_ks(L, K, P, w, r):
    return P*prod(K,L) - w*L-r*K
profitt_ks(L,K, P, w, r)
```

```python
# maksimer profitten mhp arbeidskraft (L)
dpi_L=sp.diff(profitt_ks(L,K, P, w, r),L)
foc_L=sp.Eq(dpi_L,0)
foc_L
```

```python
# Løs for L
L_ks=sp.solve(foc_L,L)[0]
L_ks
```

```python
L_sol=L_ks.subs([(P,10),(K,1)])
L_sol
```

```python

def mrp(L,K, P):
    return (.5*P*K**.5)/L**.5
mrp(L,K, P)
```
