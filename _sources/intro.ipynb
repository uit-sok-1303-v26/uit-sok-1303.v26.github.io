## Velkommen til dette kurset i Mikroøkonomi!

Kurset skal gi en grunnleggende forståelse av hvordan individer og bedrifter tar beslutninger, hvordan markeder fungerer, og hvordan ressurser fordeles i samfunnet.

## Pensum

I dette kurset bruker vi læreboken Mikroøkonomi – en ABC på 1-2-3 av Kjetil Bjorvatn (2. utgave) samt den tilhørende oppgaveboken Mikroøkonomi – en ABC på 1-2-3: Oppgaver, tips og løsningsforslag av samme forfatter. Pensum for kurset inkluderer kapittel 1–11.

## Informasjon om undervisningen i kurset

Undervisningen i dette kurset består av følgende aktiviteter:

- **Forelesninger:**
- **Seminarer:** med fokus på casearbeid, der Python benyttes for å løse oppgaver.
- **Veiledning av arbeidskrav**
- **Videoforelesninger**

## Seminarer

I dette faget er det lagt opp til seminarundervisning etter at forelesningene er gjennomført. På seminarene vil **Python-koder**, som er lagt ut i Jupyter-notatblokker, bli gjennomgått. I tillegg vil seminarene dekke anbefalte seminaroppgaver som dere finner under modulene. Det vil også bli gitt veiledning i koding knuttet til arbeidskravet.
