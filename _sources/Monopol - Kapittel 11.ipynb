### Grunnleggende monopoltilpasning
Monopolisten står overfor en etterspørselskurve og maksimerer profitt ved å sette grenseinntekten lik grensekostnaden (MR = MC).


Forklaring av variabler i monopolmodellen:

P er prisen monopolisten setter for produktet sitt.

Q er mengden av produktet som produseres totalt.

MR er grenseinntekt: Inntekten monopolisten får ved å selge én ekstra enhet.

MC er grensekostnad: Kostnaden for å produsere én ekstra enhet.

## Grenseinntekt:
I en monopolistisk setting er inntekt R(Q) definert som produktet av pris P(Q) gange med kvantum Q
$$R(Q)=P(Q)Q$$
Grenseinntekten (MR) er definert som den deriverte av inntekten med hensyn til Q:

$$MR(Q)=\frac{{\partial }R}{{\partial}Q}=P+Q\frac{{\partial }P}{{\partial}Q}$$


### Eksempel grenseinntekt når vi har en lineær etterspørselsfunksjon:

Vi skal se på den lineære etterspørselsfunksjon:

$$Q=a-bQ$$

Denne kan skrives på invers form som:

$$P=\frac{a-Q}{b}$$

Helningen på etterspørselskurven vil være lik:

$$\frac{{\partial }P}{{\partial}Q}=-\frac{1}{b}$$

Vi sette den linverse etterspørselsfunksjone inn i grenseinntekten og får:
$$MR=\frac{{\partial }R}{{\partial}Q}=P+Q\frac{{\partial }P}{{\partial}Q}=\frac{a-Q}{b}-\frac{Q}{b}=\frac{a-2Q}{b}$$


Helningen på grenseinntekstskurven er gitt ved:
$$\frac{{\partial }MR}{{\partial}Q}=-\frac{1}{b}$$


Når Q øker, reduseres marginalinntekten med en hastighet på $\frac{2}{b}$ per enhet økning i Q.

Marginalinntekten faller raskere enn prisen, noe som er en viktig egenskap i en monopolistisk markedsstruktur.

## Monopoltilpasning
Matteboks 11.1

Etterspørselsfunksjoene er gitt ved:

$$Q=a-bQ$$

Denne kan skrives på invers form som:

$$P=\frac{a-Q}{b}$$

Dette gir monopolisten inntekt lik: 

$$R(Q)=\frac{a-Q}{b}Q$$

Grenseinntekten er gitt ved: 

$$MR=\frac{{\partial }R}{{\partial}Q}=\frac{a-2Q}{b}$$

Vi antar at konstnadsfunksjonen er gitt ved: 

$$C(Q)=cQ$$

Som gir grensekostnader lik: 

$$MC=\frac{{\partial}C}{{\partial}Q}=c$$

Profittmaksimun er gitt ved: 

$$MR=MC=\frac{a-2Q}{b}=c$$

Optimalt monopolkvantum finner vi ved å løse profittmaksimun med hensyn på kvantum: 

$$Q^m=\frac{a-bC}{2}$$

Monopolprisen finner vi ved å sette optimal monopolkvantum inn i etterspørselsfunksjonen: 


$$P^m=\frac{a-Q^M}{b}=\frac{1}{2}(\frac{a}{b}+c)$$

```python
# Definerer symboler
a,b,Q,P, F,c,d,q = sp.symbols('a b Q P F c d q', real=True, positive=True)

# Definerer funksjonen for etterspørselen
P_eq = (a - Q)/b
```

```python
# For å finne inntekten, multipliserer vi prisen med kvantumet
R = P_eq*Q
R
```

```python
# Deriverer inntekten med hensyn på kvantumet og finner grenseinntekten

MR = sp.diff(R, Q)
MR
```

```python
# Definerer kostnadsfunksjonen 
def C(Q):
    return c*Q
```

```python
# Deriverer kostnadsfunksjonen med hensyn på kvantumet og finner grensekostnaden
MC = sp.diff(C(Q), Q)
MC
```

```python
# Setter MR = MC for å finne likevektskvantumet
likevekt = sp.Eq(MR, MC)
likevekt 
```

```python
# Finner optimalt monoplokavantum ved å løse for Q
Q_løsning = sp.solve(likevekt, Q)
Q_løsning
```

```python
# Setter inn kvantumet i etterspørselsfunksjonen for å finne prisen
P_løsning = P_eq = (a - Q)/b.subs(Q, Q_løsning[0])
P_løsning 
```

```python
# Setter inn kvantumet i etterspørselsfunksjonen for å finne prisen
P_løsning = P_eq.subs(Q, Q_løsning[0])
P_løsning
```
