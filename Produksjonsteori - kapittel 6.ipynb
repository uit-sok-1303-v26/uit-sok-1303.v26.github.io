### 1. Kostnader på kort sikt

Vi skal skrive kostnadsfunksjonen på kort sikt som en funksjon av hvor mange enheter bedriften produserer. På kort sikt antar vi at kapitalen ligger fast.       

 
Vi får da følgende produktfunksjon:


\begin{equation}
    x = f(N,K)=N^{0.5}K_0^{0.5}\tag{1}
\end{equation}

På kort sikt har bedriften tilgang til begrenset kapital $K_0=10$. Prisen per enhet kapital er fast $r$, mens én time med arbeidskraft koster $w$. På kort sikt kan bedriften velge hvor mange timer arbeidskraft som tilsettes. I tillegg antar vi at bridften har kostnader til som vi kaller $(Z)$. Se matteboks 6.1



```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

```python
L, K, Q, Z, K_0, w, r, z = sp.symbols('L K Q Z K_0 w r z', real=True, positive=True)
```

```python
# definer produktfunksjonen

def q(K,L):
    return (K_0**.5)*(L**.5)
```

```python
# kostnad på kort sikt
def C_s(K,L):
    return w*L+r*K_0+Z
```

```python
# Viser hvor mye arbeidskraft (L) som trenges for å produsere Q enheter ferdigprodukt

L_sol=sp.solve (q(K,L)-Q,L)[0]
L_sol
```

```python
# totale kostnad på kort sikt

def TC_s(K,L):
    return (w*Q**2/K_0)+z*Q+r*K_0
TC_s(K,L)    
   
```

```python
# varialbe kostnad på kort sikt

def VC_s(K,L):
    return (w*Q**2/K_0)+z*Q
VC_s(K,L)    
   
```

```python
# Grensekostnader
def MC(TQ,K):
    return sp.diff(VC_s(Q,K),Q)
display(MC(Q,K))
```

```python
# Gjennomsnittlige variable kostnader

def AVC_var(Q,K):
    return (VC_s(Q,K))/Q
display (AVC_var(Q,K))
```

```python
# Faste kostnad på kort sikt

def FC_s(K,L):
    return r*K_0
FC_s(K,L)   
```

```python
# Gjennomsnittlige faste kostnader

def AFC_var(Q,K):
    return (FC_s(Q,K))/Q
display (AFC_var(Q,K))
```

```python
# Gjennomsnittlige totale kostnader

def ATC_var(Q,K):
    return (TC_s(Q,K))/Q
display (ATC_var(Q,K))
```
